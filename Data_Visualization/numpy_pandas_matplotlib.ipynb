```python
import pandas as pd
import numpy as np

data = {
    "Marks": [90, np.nan, 80]
}

df = pd.DataFrame(data)

print(df.isnull().sum())
```

matplotlib


```python
import matplotlib.pyplot as plt

x = [1,2,3,4,5]
y = [10,14,12,18,20]

plt.plot(x, y, marker='o')

plt.title("Simple Line Chart")

plt.xlabel("Time")

plt.ylabel("Distance")

plt.grid(True)

plt.show()
```

```python
students = ["A","B","C","D","E"]

marks = [85,70,90,65,95]

plt.bar(students, marks)

plt.title("Student Marks")

plt.xlabel("Students")

plt.ylabel("Marks")

plt.show()
```

```python
students = ["A","B","C","D","E"]

marks = [85,70,90,65,95]

plt.bar(
    students,
    marks,
    color=["red","blue","green","orange","purple"]
)

plt.title("Student Marks")

plt.xlabel("Students")

plt.ylabel("Marks")

plt.show()
```

```python
activities = ["Study","Exercise","Sleep","Hobbies"]

hours = [4,1,7,2]

plt.pie(hours, labels=activities)

plt.show()
```
