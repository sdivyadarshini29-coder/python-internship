```python
try:
  num=int(input("Enter a value"))
  print("number:",num)
except ValueError:
  print("invalid value please try again later!")
```

```python
class Error(Exception):
    pass

u = "admin"
p = "123"

for i in range(3):
    user = input()
    pwd = input()

    if user == u and pwd == p:
        print("Success")
        break
else:
    raise Error("Login Failed")
```
