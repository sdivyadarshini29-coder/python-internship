```python
import random

# small word list
words = ["apple", "mango", "grape"]

# choose word
word = random.choice(words)

guessed = []   # store guessed letters
wrong = 0      # wrong attempts

# simple hangman stages
stages = [
"""
-----
|   |
    |
    |
    |
    |
=====
""",
"""
-----
|   |
O   |
    |
    |
    |
=====
""",
"""
-----
|   |
O   |
|   |
    |
    |
=====
""",
"""
-----
|   |
O   |
/|  |
    |
    |
=====
""",
"""
-----
|   |
O   |
/|\ |
    |
    |
=====
""",
"""
-----
|   |
O   |
/|\ |
/   |
    |
=====
""",
"""
-----
|   |
O   |
/|\ |
/ \ |
    |
=====
"""
]

print("Welcome to Hangman Game")

# game loop
while wrong < 6:

    # show word
    display = ""
    for letter in word:
        if letter in guessed:
            display = display + letter + " "
        else:
            display = display + "_ "

    print("\nWord:", display)

    # check win
    if "_" not in display:
        print("You Win 🎉")
        break

    # user input
    guess = input("Enter letter: ")

    # check repeat
    if guess in guessed:
        print("Already guessed")
        continue

    guessed.append(guess)

    # check correct/wrong
    if guess in word:
        print("Correct")
    else:
        wrong = wrong + 1
        print(stages[wrong])
        print("Wrong guess")

# lose condition
if wrong == 6:
    print("Game Over 💀")
    print("Correct word is:", word)
```
