```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv("Teen_Mental_Health_Dataset.csv")
```

```python
missing=df.isnull().sum()
```

```python
platform_usage_counts=df['platform_usage'].value_counts()
print("platform usage counts:\n",platform_usage_counts)
```

```python
sns.countplot(x='platform_usage', data=df, palette='pastel')
plt.title('platform usage counts')
plt.xlabel('platform usage')
plt.ylabel('count')
plt.show()
```

```python
gender_counts = df['gender'].value_counts()
print("\nGender Counts:\n", gender_counts)
```

```python
gender_by_platform_usage = df.groupby(['gender', 'platform_usage']).size()
print("\ngender by platform usage:\n", gender_by_platform_usage)
```

```python
sns.countplot(x='gender', hue='platform_usage', data=df, palette='Set2')
plt.title('gender by platform usage')
plt.xlabel('gender')
plt.ylabel('count')
plt.show()
```

```python
plt.figure(figsize=(6,5))

plt.boxplot(
    df['academic_performance'],
    patch_artist=True,
    boxprops=dict(facecolor='violet')
)

plt.title("Academic Performance")
plt.ylabel("Performance")

plt.show()
```

```python
plt.figure(figsize=(7,5))

sns.histplot(
    df['sleep_hours'],
    bins=10,
    color='orange',
    kde=True
)

plt.title("Sleep Hours Distribution")

plt.show()
```

```python
plt.figure(figsize=(7,5))

sns.scatterplot(
    x='daily_social_media_hours',
    y='stress_level',
    hue='gender',
    data=df,
    palette='Set1'
)

plt.title("Social Media vs Stress")

plt.show()
```

```python
plt.figure(figsize=(12,8))

numeric_df = df.select_dtypes(include=['int64', 'float64'])

sns.heatmap(
    numeric_df.corr(),
    annot=True,
    cmap='coolwarm'
)

plt.title("Correlation Heatmap")

plt.show()
```

```python
sns.kdeplot(df['age'], fill=True, color='skyblue')
plt.title('Age Distribution (KDE Plot)')
plt.xlabel('Age')
plt.ylabel('Density')
plt.show()
```
